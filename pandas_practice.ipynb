# Merge method

```python
# https://datascienceschool.net/view-notebook/7002e92653434bc88c8c026c3449d27b/
# 만약 테이블에 키 값이 같은 데이터가 여러개 있는 경우에는 있을 수 있는 모든 경우의 수를 따져서 조합을 만들어 낸다.
# 두 데이터프레임에서 이름이 같은 열은 모두 키가 된다. 만약 이름이 같아도 키가 되면 안되는 열이 있다면 on 인수로 기준열을 명시해야 한다.
# 반대로 키가 되는 기준열의 이름이 두 데이터프레임에서 다르다면 left_on, right_on 인수를 사용하여 기준열을 명시해야 한다.
# 일반 데이터 열이 아닌 인덱스를 기준열로 사용하려면 left_index 또는 right_index 인수를 True 로 설정한다.

# index 자동설정된 데이터에 대해
df1 = pd.DataFrame(data = {"timestamp": stamp, "latitude": lat, "longitude": long})
df2 = pd.DataFrame(data = {"timestamp": event_time, "message" : event_list})
pd.merge(df1, df2, how = 'inner') # default = inner join 둘다 값 있는 것만
pd.merge(df1, df2, how = 'outer') # outer join 하나만 값 있어도
pd.merge(df1, df2, how = 'left') # left key 값만
pd.merge(df1, df2, how = 'right') # right key 값만
```

```python
# index 수동설정된 데이터에 대해서
df1 = pd.DataFrame(data = {"latitude": lat, "longitude": long}, index = stamp)
df1.index.name = "timestamp"
df2 = pd.DataFrame(data = {"message" : event_list}, index = event_time)
df2.index.name = "timestamp"
pd.merge(df1, df2, left_on = "timestamp" ,right_index= True) # default = inner join
pd.merge(df1, df2, right_on = "timestamp" ,left_index= True) 
pd.merge(df1, df2, right_index= True, left_index= True)
```

```python
class Listdataset(object):
    def __init__(self, pos, msg, mode, temp, mile):
        """
        [lat_list, long_list, loc_stamp]
        [event_list, event_time]
        [mode_list, mode_stamp]
        [CPU_list, GPU_list, temp_stamp]
        [mile_list, mile_stamp]
        """
        self.pos = pos
        self.msg = msg
        self.mode = mode
        self.temp = temp
        self.mile = mile

    def printpos(self,num = 5):
        print("timestamp_sample = {}".format(self.pos[2][:num]))
        print("lat_list_sample = {}".format(self.pos[0][:num]))
        print("long_list_sample = {}".format(self.pos[1][:num]))

    def printmsg(self,num = 5):
        print("timestamp_sample = {}".format(self.msg[1][:num]))
        print("event_list_sample = {}".format(self.msg[0][:num]))

    def printmode(self, num = 5):
        print("timestamp_sample = {}".format(self.mode[1][:num]))
        print("mode_list_sample = {}".format(self.mode[0][:num]))

    def printtemp(self, num = 5):
        print("timestamp_sample = {}".format(self.temp[2][:num]))
        print("CPU_list_sample = {}".format(self.temp[0][:num]))
        print("GPU_list_sample = {}".format(self.temp[1][:num]))

    def printmile(self, num = 5):
        print("timestamp_sample = {}".format(self.mile[1][:num]))
        print("mile_list_sample = {}".format(self.mile[0][:num]))

    def printsample(self, num=5):
        try:
            self.printpos(num)
            self.printmsg(num)
            self.printmode(num)
            self.printtemp(num)
            self.printmile(num)
        except IndexError:
            print("we need at least 5 length samples\n\
                   or your number is larger than the number of samples")
```

```python
def df_to_ldset(df):
    """
    This function is to transform DataFrame type into Listdataset type
    """
    df_pos = df.loc[:,df.columns[0:2]].dropna()
    df_msg = df.loc[:,df.columns[2]].dropna()
    df_mode = df.loc[:,df.columns[3]].dropna()
    df_temp = df.loc[:,df.columns[4:6]].dropna()
    df_mile = df.loc[:,df.columns[6]].dropna()
    ldset = Listdataset(
        [df_pos.iloc[:,0].values, df_pos.iloc[:,1].values, df_pos.index.values],
        [df_msg.values, df_msg.index.values],
        [df_mode.values, df_mode.index.values],
        [df_temp.iloc[:,0].values, df_temp.iloc[:,1].values, df_temp.index.values],
        [df_mile.values, df_mile.index.values])
    return ldset

def ldset_to_df(ldset):
    """
    This function is to transfrom Listdataset type into DataFrame type
    """
    [lat_list, long_list, loc_stamp] = ldset.pos
    [event_list, event_time] = ldset.msg
    [mode_list, mode_stamp] = ldset.mode
    [CPU_list, GPU_list, temp_stamp] = ldset.temp
    [mile_list, mile_stamp] = ldset.mile

    # set DataFrame
    df1 = pd.DataFrame(data = {"latitude": lat_list, "longitude": long_list}, index = loc_stamp)
    df1.index.name = "timestamp"
    df2 = pd.DataFrame(data = {"message" : event_list}, index = event_time)
    df2.index.name = "timestamp"
    df3 = pd.DataFrame(data = {"automode" : mode_list}, index = mode_stamp)
    df3.index.name = "timestamp"
    df4 = pd.DataFrame(data = {"cpu_temp": CPU_list, "gpu_temp": GPU_list}, index = temp_stamp)
    df4.index.name = "timestamp"
    df5 = pd.DataFrame(data = {"mile": mile_list}, index = mile_stamp)
    df5.index.name = "timestamp"

    # merge
    df = pd.merge(df1, df2, right_index= True, left_index= True, how = 'outer')
    df = pd.merge(df, df3, right_index= True, left_index= True, how = 'outer')
    df = pd.merge(df, df4, right_index= True, left_index= True, how = 'outer')
    df = pd.merge(df, df5, right_index= True, left_index= True, how = 'outer')
    return df
```

#### 몇시각전의 data를 보여주기
timedelta사용하여 이전시점의 timestamp값 찾고  
그 표에서 가장 가까운 timestamp값을 찾아서 저장  
이전시점 중 가장가까운timestamp와 message timestamp의 값으로 슬라이싱  
가장가까운 값은 슬라이싱을 하면 pandas모듈에서 알아서 예외처리하여 찾아준다. 

[timestamp를 datetime으로 혹은 datetime을 timestamp로 변환하기](https://ourcstory.tistory.com/109)  

```python
import time
```

```python
msg_time = 1565145650
delta_time = 10
dtt = datetime.datetime.fromtimestamp(msg_time)
dtt_delta = datetime.timedelta(seconds=delta_time)
dtt_minus = dtt - dtt_delta
prev_time = time.mktime(dtt_minus.timetuple())
df.loc[prev_time:msg_time]
```

#### 두개의 특정시점을 받아와 슬라이싱하기

```python
year = 2019
month = 8
day = 7
hour = 11
minute = 40
second1 = 40
second2 = 51
dtt1 = datetime.datetime(year=year, month=month, day=day, hour=hour, minute=minute, second= second1)
dtt2 = datetime.datetime(year=year, month=month, day=day, hour=hour, minute=minute, second= second2)
time1 = time.mktime(dtt1.timetuple())
time2 = time.mktime(dtt2.timetuple())
df.loc[time1:time2]
```
